```python

```

Please run the above codes and print the values for A in both two code blocks for comparison. In your comments, explain why outputs are different.

````python
## Put your codes/comments and explain the results in this cell
print('Q1:')

A = ['apple', 'orange']
print(A)
B = A
print(id(A),id(B))

B[0] = 'grape'
print(A)

# it change the output of A.
# The reason is A is a list and A[0] is apple, A[1] is orange
# by B=A, B point to A's address. Thus, change B[0]= 'grape' also changed A[0]


print('Q2:')
A = ['apple', 'orange']
print(A)
B = A[:]
print(id(A),id(B))
B[0] = 'grape'
print(A)
# it not change the output of A
# The reason is B = A[:] will gives B a new address, Thus change B[0] won't change A




````
